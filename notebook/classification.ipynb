```python
import re
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer=WordNetLemmatizer()
```

```python
stopwords_list=set(stopwords.words('english'))
```

```python
stopwords_list
```

```python
from nltk.corpus import wordnet
def get_wordnet_pos(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
corpus=[]

for i in range(len(df)):
    cleaned=re.sub(r'[^a-zA-Z]',' ',df['text'].iloc[i])

    cleaned=cleaned.lower()

    cleaned=cleaned.split()
    
    pos_tags=nltk.pos_tag(cleaned)

    cleaned=[lemmatizer.lemmatize(word,pos=get_wordnet_pos(pos)) for word,pos in pos_tags if word not in stopwords_list]
    cleaned=' '.join(cleaned)

    corpus.append(cleaned)

    
```

```python
corpus
```

```python
#encoding
# harassment->1
# normal->0

df['label']=df['label'].map({'harassment':1,'normal':0})
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20,random_state=42)
```

```python
models = {
    "MultinomialNB": MultinomialNB(),
    "LogisticRegression": LogisticRegression(),
}
```

```python
from sklearn.model_selection import GridSearchCV

```

```python
params = {
    "MultinomialNB": [
        {
            'alpha': [0.1, 0.5, 1.0, 1.5, 2.0],   # smoothing parameter
            'fit_prior': [True, False]           # whether to learn class prior probabilities
        }
    ],
    
    "LogisticRegression": [
        {
            'C': [0.01, 0.1, 1, 10],             # inverse of regularization strength
            'penalty': ['l1', 'l2'],             # L1 = Lasso, L2 = Ridge
            'solver': ['liblinear'],             # solver compatible with L1 & L2
            'max_iter': [500]                     # ensure convergence
        }
    ]
}
```

```python
params
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

for name, model in models.items():
    print(f"\n{name} training...")

    # Train the model
    grid=GridSearchCV(estimator=model,param_grid=params[name],cv=5,scoring='f1',n_jobs=-1)

    grid.fit(X_train,y_train)

    # Predictions
    y_train_pred = grid.predict(X_train)
    y_test_pred = grid.predict(X_test)

    # Training metrics
    print("\n--- Training Metrics ---")
    print("Accuracy:", accuracy_score(y_train, y_train_pred))
    print("Precision:", precision_score(y_train, y_train_pred, pos_label=1))
    print("Recall:", recall_score(y_train, y_train_pred, pos_label=1))
    print("F1-score:", f1_score(y_train, y_train_pred, pos_label=1))

    # Testing metrics
    print("\n--- Testing Metrics ---")
    print("Accuracy:", accuracy_score(y_test, y_test_pred))
    print("Precision:", precision_score(y_test, y_test_pred, pos_label=1))
    print("Recall:", recall_score(y_test, y_test_pred, pos_label=1))
    print("F1-score:", f1_score(y_test, y_test_pred, pos_label=1))

    print()
    print("best params:",grid.best_params_)
    print("--"*20)
    print()
```

```python
def clean_message(text):
    cleaned=re.sub(r'[^a-zA-Z]',' ',text)

    cleaned=cleaned.lower()

    cleaned=cleaned.split()
    
    pos_tags=nltk.pos_tag(cleaned)

    cleaned=[lemmatizer.lemmatize(word,pos=get_wordnet_pos(pos)) for word,pos in pos_tags if word not in stopwords_list]
    cleaned=' '.join(cleaned)

    return cleaned
```

```python
text = "The cats are running faster than the dogs!"
clean_message(text)
```

```python
def predict(text):
    cleaned=clean_message(text)
    X_new=tfidf.transform([cleaned])
    pred=model.predict(X_new)

    if pred[0]==1:
        return 'harrasment'
    else:
        return 'normal'
```
